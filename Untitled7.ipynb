```python
# 2. local functions

# extract sequence and ss from the dssp tuple
def get_AA_SS_from_DSSP(dssp_1_s1):
  AA_Full = np.array([]) # ''
  SS_Full = np.array([]) # ''
  ChainID_Full = np.array([]) # ''

  for z in range(len(dssp_1_s1)):
    a_key = list(dssp_1_s1.keys())[z]
    # ChainID_Full += a_key[0]
    # AA_Full += dssp_1_s1[a_key][1]
    # SS_Full += dssp_1_s1[a_key][2]
    # for np
    ChainID_Full = np.append( ChainID_Full,a_key[0] )
    AA_Full = np.append( AA_Full, dssp_1_s1[a_key][1] )
    SS_Full = np.append( SS_Full, dssp_1_s1[a_key][2] )
  return ChainID_Full,AA_Full,SS_Full

# divide the result records according to the chains
def Divide_Result_For_Chains(ChainID_Full,AA_Full,SS_Full):
  UniChainList = ''.join(set(ChainID_Full))
  NChain=len(UniChainList)

  resu={"Chain":[],"AA":[],"SS":[]}
  for ThisChain in UniChainList:
    id_pick=ChainID_Full==ThisChain
    This_AA=AA_Full[id_pick]
    This_SS=SS_Full[id_pick]
    # debug
    # print(This_AA)
    resu["Chain"].append(ThisChain)
    # resu["AA"].append(["".join(item) for item in This_AA.astype(str)])
    resu["AA"].append(''.join(map(str, This_AA)))
    resu["SS"].append(''.join(map(str, This_SS)))
  
  return NChain,resu

def calculate_SS_using_DSSP(pdb_name,pdb_file):
  p = PDBParser()
  structure = p.get_structure(pdb_name, pdb_file)
  model = structure[0]

  dssp = DSSP(model, pdb_file)
  # dssp = DSSP_SelfMade(model, pdb_file)

  ChainID_Full,AA_Full,SS_Full=get_AA_SS_from_DSSP(dssp)
  NChain, Resu_Dic = \
  Divide_Result_For_Chains(ChainID_Full,AA_Full,SS_Full)

  return NChain, Resu_Dic
```

```python
# print the results:
def print_result_to_file(ou_file,pdb_name,NChain,Resu_Dic):
  open_file = 0
  with open(ou_file, 'w') as f:
    open_file = 1
    for i_reco in range(NChain):
      reco_line=pdb_name+','+Resu_Dic["Chain"][i_reco]+\
                ','+Resu_Dic["AA"][i_reco]+ \
                ','+Resu_Dic["SS"][i_reco]
      f.write(reco_line+'\n')
  return open_file
```

```python
# extract sequence and ss from the dssp tuple
def get_AA_SS_from_DSSP_SelfMade(dssp_1_s1):
  AA_Full = np.array([]) # ''
  SS_Full = np.array([]) # ''
  ChainID_Full = np.array([]) # ''
  # for the extended
  PHI_Full = np.array([])
  PSI_Full = np.array([])
  # useing self-made module
  KAPPA_Full = np.array([])
  ALPHA_Full = np.array([])
  CAX_Full = np.array([])
  CAY_Full = np.array([])
  CAZ_Full = np.array([])


  for z in range(len(dssp_1_s1)):
    a_key = list(dssp_1_s1.keys())[z]
    # ChainID_Full += a_key[0]
    # AA_Full += dssp_1_s1[a_key][1]
    # SS_Full += dssp_1_s1[a_key][2]
    # for debug
    # print( dssp_1_s1[a_key] )

    # for np
    ChainID_Full = np.append( ChainID_Full,a_key[0] )
    AA_Full = np.append( AA_Full, dssp_1_s1[a_key][1] )
    SS_Full = np.append( SS_Full, dssp_1_s1[a_key][2] )
    # for the extended
    PHI_Full = np.append( PHI_Full, dssp_1_s1[a_key][4] )
    PSI_Full = np.append( PSI_Full, dssp_1_s1[a_key][5] )
    # using sel-made module
    KAPPA_Full = np.append( KAPPA_Full, dssp_1_s1[a_key][14] )
    ALPHA_Full = np.append( ALPHA_Full, dssp_1_s1[a_key][15] )
    CAX_Full = np.append( CAX_Full, dssp_1_s1[a_key][16] )
    CAY_Full = np.append( CAY_Full, dssp_1_s1[a_key][17] )
    CAZ_Full = np.append( CAZ_Full, dssp_1_s1[a_key][18] )

  return ChainID_Full,AA_Full,SS_Full,PHI_Full,PSI_Full, \
         KAPPA_Full,ALPHA_Full,CAX_Full,CAY_Full,CAZ_Full

# divide the result records according to the chains
def Divide_Result_For_Chains_SelfMade(ChainID_Full,AA_Full,SS_Full, \
                                      PHI_Full,PSI_Full,\
                                      KAPPA_Full,ALPHA_Full, \
                                      CAX_Full,CAY_Full,CAZ_Full):
  UniChainList = ''.join(set(ChainID_Full))
  NChain=len(UniChainList)

  resu={"Chain":[],"AA":[],"SS":[],"PHI":[],"PSI":[],\
        "KAPPA":[],"ALPHA":[],"CAX":[],"CAY":[],"CAZ":[]}
  for ThisChain in UniChainList:
    id_pick=ChainID_Full==ThisChain
    This_AA=AA_Full[id_pick]
    This_SS=SS_Full[id_pick]
    # +++++++++++++++++++++++++
    This_PHI=PHI_Full[id_pick]
    This_PSI=PSI_Full[id_pick]
    This_KAPPA=KAPPA_Full[id_pick]
    This_ALPHA=ALPHA_Full[id_pick]
    This_CAX=CAX_Full[id_pick]
    This_CAY=CAY_Full[id_pick]
    This_CAZ=CAZ_Full[id_pick]


    # debug
    # print(This_AA)
    resu["Chain"].append(ThisChain)
    # resu["AA"].append(["".join(item) for item in This_AA.astype(str)])
    resu["AA"].append(''.join(map(str, This_AA)))
    resu["SS"].append(''.join(map(str, This_SS)))
    # ++++++++++++++++++++++++++++
    resu["PHI"].append(' '.join(map(str, This_PHI)))
    resu["PSI"].append(' '.join(map(str, This_PSI)))
    resu["KAPPA"].append(' '.join(map(str, This_KAPPA)))
    resu["ALPHA"].append(' '.join(map(str, This_ALPHA)))
    resu["CAX"].append(' '.join(map(str, This_CAX)))
    resu["CAY"].append(' '.join(map(str, This_CAY)))
    resu["CAZ"].append(' '.join(map(str, This_CAZ)))
  
  return NChain,resu

def calculate_SS_using_DSSP_SelfMade(pdb_name,pdb_file):
  p = PDBParser()
  structure = p.get_structure(pdb_name, pdb_file)
  model = structure[0]

  # dssp = DSSP(model, pdb_file)
  dssp = DSSP_SelfMade(model, pdb_file)

  ChainID_Full,AA_Full,SS_Full,PHI_Full,PSI_Full, \
         KAPPA_Full,ALPHA_Full, \
         CAX_Full,CAY_Full,CAZ_Full=get_AA_SS_from_DSSP_SelfMade(dssp)
  NChain, Resu_Dic = \
  Divide_Result_For_Chains_SelfMade(ChainID_Full,AA_Full,SS_Full,\
                                    PHI_Full,PSI_Full, \
                                    KAPPA_Full,ALPHA_Full, \
                                    CAX_Full,CAY_Full,CAZ_Full)

  return NChain, Resu_Dic

# print the results:
def print_result_to_file_SelfMade(ou_file,pdb_name,NChain,Resu_Dic):
  open_file = 0
  with open(ou_file, 'w') as f:
    open_file = 1
    for i_reco in range(NChain):
      reco_line=pdb_name+','+Resu_Dic["Chain"][i_reco]+\
                ','+Resu_Dic["AA"][i_reco]+ \
                ','+Resu_Dic["SS"][i_reco]+ \
                ','+Resu_Dic["PHI"][i_reco]+ \
                ','+Resu_Dic["PSI"][i_reco]+ \
                ','+Resu_Dic["KAPPA"][i_reco]+ \
                ','+Resu_Dic["ALPHA"][i_reco]+ \
                ','+Resu_Dic["CAX"][i_reco]+ \
                ','+Resu_Dic["CAY"][i_reco]+ \
                ','+Resu_Dic["CAZ"][i_reco]
      f.write(reco_line+'\n')

  return open_file
```
